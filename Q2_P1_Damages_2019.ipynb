```python
# Creating dataset for damaged land parts of the sates
land_destroyed= forest.filter(["State","Conservation", 
                               "Forests/Plantation", 
                               "Agriculture", "Total_damaged"])

land_destroyed
```

# Visualization

```python
import matplotlib as mpl
#plt.style.use('seaborn-white')
```

## 1. Total deaths

```python
plt.figure(figsize=(6,4), dpi= 80)
my_plot = sns.barplot(x="State", y="Fatalitites", data=forest)

plt.xlabel("States")
plt.ylabel("Total deaths")
plt.title("Black Summer: Total death count")

# save image
fig= plt.gcf()
fig.savefig("total_deaths.png")

plt.tight_layout()
```

## 2. Total land destroyed

```python
plt.figure(figsize=(6,4), dpi= 80)
my_plot = sns.barplot(x="state", y="Area", data=forest2)

plt.xlabel("State")
plt.ylabel("Area in Hectares")
plt.title("BushFire damage to Agricultural lands/forests")

# save image
fig= plt.gcf()
fig.savefig("total_land_damaged.png")

plt.tight_layout()
```

## 3. Total houses burnt

```python
plt.figure(figsize=(5,4), dpi= 80)
my_plot = sns.barplot(x="State", y="Homes Lost", data=forest)

plt.xlabel("States")
plt.ylabel("Total houses lost")
plt.title("BushFire damage to Houses")

# save image
fig= plt.gcf()
fig.savefig("homes_lost.png")

plt.tight_layout()
```

## 4. total land damage in all states

```python
land_tidy= land_destroyed.melt(id_vars="State", 
                              var_name="Land_type", 
                              value_name="Hectares")

#land_tidy
```

```python
plt.figure(figsize=(8,4), dpi= 80)
my_plot = sns.barplot(x="State", y="Hectares",hue = "Land_type", data=land_tidy)
fig= plt.gcf()
fig.savefig("land_damaged.png")


plt.tight_layout()
```
