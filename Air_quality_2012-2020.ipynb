# AIR Quality data NSW


**Data Source:**
+ https://www.dpie.nsw.gov.au/air-quality/search-for-and-download-air-quality-data

Air Quality data is taken from NSW Government data from 2012 to 2020.
**Data Preprocessing:**

The data included other paramters like CO2, CO, Ozone etc. which were removed in Excel. In the following notebook, the data is only for the air quality index. A new variable ***avg_Air_q*** is created that has an average air quality value for all regions in NSW.

```python
import os
```

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt #plotting
import seaborn as sns #for beatiful visualization
```

```python
Air_12to20= pd.read_csv(r'C:\\Users\\61423\\OneDrive\\Desktop\\SEM 3\\DVN\\AT3\\smoke-air quality\\NSW_air_quality(2012-2020).csv')
Air_12to20.head()
```

```python
from datetime import datetime, date
Air_12to20["Date"]= pd.to_datetime(Air_12to20["Date"]).dt.normalize()
Air_12to20["Date"]
```

```python
Air_12to20["Date"]= pd.to_datetime(Air_12to20["Date"])
```

```python
# Taking average of air quality index for all regions in NSW 
Air_12to20['avg_air_q'] = Air_12to20.iloc[:,1:].mean(axis=1)
Air_12to20.head()
```

```python
# Save the data into a csv
Air_12to20.to_csv("Air_Q_NSW_12TO20.csv")
```

## Convert daily data into monthly data

```python
# set Date to datetimeindex
Air_12to20.set_index('Date', inplace=True)
Air_12to20.index = pd.to_datetime(Air_12to20.index)
```

```python
Air_q_monthly = Air_12to20.resample('M').agg({'avg_air_q':'max'})
Air_q_monthly.head()
```

```python
# to bring all dates on same scale
#Air_q_monthly.index = [pd.datetime(x.year, x.month, 1) for x in Air_q_monthly.index.tolist()] 
```

```python
# Naming the index column
Air_q_monthly.index.name= "Date"

# Converting the index into a column
Air_q_month= Air_q_monthly.reset_index()
print("modified: ")
print(Air_q_month)
```

```python
# Save the motnhly Air Quality into a csv
Air_q_month.to_csv("Air_quality_monthly.csv")
```

### Barplot: Monthly Airquality(2012-2020)

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
```

```python
# Define range of dates for the barchart
mask = (Air_q_month['Date'] > '2012-01-31') & (Air_q_month['Date'] <= '2020-03-01')
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
plt.figure(figsize=(20,10), dpi= 80)
ax = sns.barplot(x='Date',y='avg_air_q',data=Air_q_month.loc[mask])



ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right") # this rotates the labels at a 40degree

plt.xlabel("Date")
plt.ylabel('Air Quality Index')
plt.title("NSW Air Quality (2012 to 2020)")
plt.tight_layout()


ticklabels = ['']*len(Air_q_month)
skip = len(Air_q_month)//18
ticklabels[::skip] = Air_q_month['Date'].iloc[::skip].dt.strftime('%Y-%m-%d')
ax.xaxis.set_major_formatter(mticker.FixedFormatter(ticklabels))
# fig.autofmt_xdate()

# fixes the tracker
# https://matplotlib.org/users/recipes.html
def fmt(x, pos=0, max_i=len(ticklabels)-1):
    i = int(x) 
    i = 0 if i < 0 else max_i if i > max_i else i
    return dates[i]
ax.fmt_xdata = fmt
plt.show()

plt.savefig('Air_quality_monthly_12to20.png')
```

## Convert monthly data into yearly

```python
Air_q_yearly = Air_12to20.resample('Y').agg({'avg_air_q':'max'})
Air_q_yearly.head()
```

```python
# Naming the index column
Air_q_yearly.index.name= "Year"

# Converting the index into a column
Air_q_year= Air_q_yearly.reset_index()
#print("modified: ")
#print(Air_q_year)
```

```python
# Define range of dates for the barchart
mask_1 = (Air_q_year['Year'] > '2012-01-31') & (Air_q_year['Year'] <= '2020-03-01')
```

```python
Air_q_year["Year"]=pd.to_datetime(Air_q_year.Year)
print(Air_q_year)
```

```python
# Save the yearly air Quality data into a csv
Air_q_year.to_csv("Air_quality_yearly.csv")
```

```python
plt.figure(figsize=(10,10), dpi= 80)
#sns.set_palette("pastel")
ax = sns.barplot(x='Year',y='avg_air_q',data=Air_q_year.loc[mask_1])

# format the coords message box
ax.format_xdata = mdates.DateFormatter('%Y-%m-%d')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right") # this rotates the labels at a 40degree

plt.xlabel("Date")
plt.ylabel('Air Quality Index')
plt.title("NSW Air Quality (2012 to 2020)")
plt.tight_layout()

plt.show()
```
