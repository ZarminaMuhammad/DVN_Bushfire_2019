```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt #plotting
import seaborn as sns #for beatiful visualization
```

```python
# reading csv
ozone_years= pd.read_csv(r'C:\\Users\\61423\\OneDrive\\Desktop\\SEM 3\\DVN\\AT3\\smoke-air quality\\ozone(NSW).csv')
ozone_years
```

```python
from datetime import datetime, date
ozone_years["Date"]= pd.to_datetime(ozone_years["Date"]).dt.normalize()
ozone_years["Date"]
```

```python
# take average of all locations 
```

```python
# Taking average of air quality index for all regions in NSW 
ozone_years['ozone_avg'] = ozone_years.iloc[:,1:].mean(axis=1)
ozone_years.head()
```

```python
# set Date to datetimeindex
ozone_years.set_index('Date', inplace=True)
ozone_years.index = pd.to_datetime(ozone_years.index)
```

```python
ozone_years_avg = ozone_years.resample('M').agg({'ozone_avg':'max'}) 
ozone_years_avg.head()
```

```python
# Naming the index column
ozone_years_avg.index.name= "Date"

# Converting the index into a column
ozone_years_avg= ozone_years_avg.reset_index()
print("modified: ")
print(ozone_years_avg)
```

```python
# Save the year+months ozone data into a csv
ozone_years_avg.to_csv("ozone_monthly_2001-2019.csv")
```

```python
# Plot ozone content
```

```python
from matplotlib import pyplot
#plt.style.use('ggplot')
plt.style.use('seaborn-white')
```

```python
ozone_years_avg.plot(x= "Date", y="ozone_avg")

fig= plt.gcf()
#fig.savefig("ozone_monthly.png")
pyplot.xlabel("Year")
pyplot.ylabel('ozone_avg')
pyplot.title("Australia(NSW): Increase in ozone (2001-2019)")
fig.savefig("ozone_monthly.png")
pyplot.show()


#plt.savefig('Aus_Diff_in_temp.png')
```

```python
# Define range of dates for the barchart
mask = (ozone_years_avg['Date'] > '2017-12-31')
```

```python
ozone_years_avg.loc[mask].plot(x= "Date", y="ozone_avg")

fig= plt.gcf()
#fig.savefig("ozone_monthly.png")
pyplot.xlabel("")
pyplot.ylabel('ozone concentration')
pyplot.title("Black Summer(NSW): Increase in ozone")
fig.savefig("ozone_2019.png")
pyplot.show()
```

```python
# conversion to yearly data
```

```python
ozone_yearly = ozone_years.resample('Y').agg({'ozone_avg':'max'})
ozone_yearly.head()
```

```python
# Naming the index column
ozone_yearly.index.name= "Date"

# Converting the index into a column
ozone_yearly= ozone_yearly.reset_index()
print("modified: ")
print(ozone_yearly)
```

```python
ozone_yearly['year'] = ozone_yearly['Date'].apply(lambda x: x.strftime('%Y'))
print(ozone_yearly)
```

```python
ozone_yearly.plot(x= "year", y="ozone_avg")

fig= plt.gcf()
#fig.savefig("ozone_yearly.png")
#pyplot.xlabel("Year")
pyplot.ylabel('ozone contration')
pyplot.title("Australia(NSW): Increase in ozone")
fig.savefig("ozone_yearly.png")
pyplot.show()
```

```python
# Save the year+months ozone data into a csv
ozone_yearly.to_csv("ozone_yearly_2001-2019.csv")
```
