```python
import matplotlib as mpl
#plt.style.use('seaborn-white')
```

```python
plt.figure(figsize=(6,4), dpi= 80)
my_plot = sns.barplot(x="Year", y="Area burned('000 acres)", data=years)

#my_plot.set_xticklabels(my_plot.get_xticklabels(), rotation=40, ha="right")
plt.xlabel("Year")
plt.ylabel("Land damage ('000 acres)")
plt.title("")

# save image
fig= plt.gcf()
fig.savefig("Land_damage_years.png")

plt.tight_layout()
```

```python
plt.figure(figsize=(6,4), dpi= 80)
my_plot = sns.barplot(x="Year", y="Properties damaged", data=years)

my_plot.set_xticklabels(my_plot.get_xticklabels(), rotation=40, ha="right")
plt.xlabel("Year")
plt.ylabel("Properties damage")
plt.title("")

# save image
fig= plt.gcf()
fig.savefig("Properties_years.png")

plt.tight_layout()
```

```python
plt.figure(figsize=(6,4), dpi= 80)
my_plot = sns.barplot(x="Bushfire_name", y="Properties damaged", data=years)

my_plot.set_xticklabels(my_plot.get_xticklabels(), rotation=40, ha="right")
plt.xlabel("Year")
plt.ylabel("Total properties damage")
plt.title("")

# save image
fig= plt.gcf()
fig.savefig("Properties-damage_years.png")

plt.tight_layout()
```
